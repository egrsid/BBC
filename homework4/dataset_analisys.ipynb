средний возраст выживших и погибших мужчин и женщин

```python
mam = df['Age'][df['Sex'] == 'male'][df['Survived'] == 1].mean()
mdm = df['Age'][df['Sex'] == 'male'][df['Survived'] == 0].mean()

wam = df['Age'][df['Sex'] == 'female'][df['Survived'] == 1].mean()
wdm = df['Age'][df['Sex'] == 'female'][df['Survived'] == 0].mean()

mam, mdm, wam, wdm  # тк все мужчины умерли, а все мужчины выжили
```
